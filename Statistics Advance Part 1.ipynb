```python
# Required libraries for statistical analysis and visualization

import numpy as np                    # For numerical operations and generating random data
import matplotlib.pyplot as plt       # For plotting graphs
import seaborn as sns                 # For enhanced visualizations (like KDE plots and histograms)

from scipy.stats import (             # Importing specific statistical functions
    norm,                             # Normal distribution (PDF, CDF)
    binom,                            # Binomial distribution
    poisson,                          # Poisson distribution
    uniform,                          # Uniform distribution
    zscore,                           # Z-score normalization
    ttest_1samp,                      # One-sample t-test
    kurtosis,                         # Kurtosis calculation
    skew,                             # Skewness calculation
    sem                               # Standard error of the mean
)
import scipy.stats as stats           # For confidence intervals and t-distribution operations
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete values (e.g., rolling a fair 6-sided die)
x = np.arange(1, 7)
pmf = np.ones_like(x) / len(x)  # Equal probability for each outcome

plt.stem(x, pmf, basefmt=" ")
plt.title("Discrete Uniform Distribution PMF")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
# Question: What is the probability of getting exactly 5 successes in 10 trials with p = 0.5?

from scipy.stats import binom

n, p = 10, 0.5
k = 5
prob = binom.pmf(k, n, p)
print(f"Probability of exactly {k} successes: {prob}")
```

```python
# Question: What is the Z-score for a data point and how does it relate to the standard normal distribution?

data_point = 72
mean = 70
std_dev = 5
z_score = (data_point - mean) / std_dev
print(f"Z-score: {z_score}")
```

```python
# Question: Is the sample mean significantly different from the population mean using Z-test?

sample = np.random.normal(52, 10, 30)
pop_mean = 50
z_stat = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
print(f"Z-statistic: {z_stat}")
```

```python
# Question: What is the 95% confidence interval for the mean of a dataset?

import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
print(f"95% Confidence Interval: {ci}")
```

```python
# Question: How can we compute a confidence interval for a normally distributed sample?

data = np.random.normal(60, 8, 100)
mean = np.mean(data)
sem = stats.sem(data)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
print(f"Confidence Interval for Mean: {conf_interval}")
```

```python
# Question: What is the cumulative probability of up to 4 events in a Poisson distribution with λ = 3?

from scipy.stats import poisson

lambda_val = 3
k = 4
cdf_val = poisson.cdf(k, mu=lambda_val)
print(f"CDF (k ≤ {k}) for Poisson(λ=3): {cdf_val}")
```

```python
# Question: What is the expected value of a uniform distribution from a to b?

a, b = 10, 20
uniform_samples = np.random.uniform(a, b, 1000)
expected_val = np.mean(uniform_samples)
print(f"Expected value (mean): {expected_val}")
```

```python
# Question: How do the spreads (standard deviations) of two datasets compare?

data1 = np.random.normal(100, 5, 1000)
data2 = np.random.normal(100, 20, 1000)

std1 = np.std(data1)
std2 = np.std(data2)
print(f"Standard Deviation of Dataset 1: {std1}")
print(f"Standard Deviation of Dataset 2: {std2}")

plt.figure()
sns.kdeplot(data1, label='Low Std Dev')
sns.kdeplot(data2, label='High Std Dev')
plt.title("Comparison of Standard Deviations")
plt.legend()
plt.show()
```

```python
# Question: What is the range and interquartile range (IQR) of a dataset?

data = np.random.normal(70, 15, 500)
range_val = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
print(f"Range: {range_val}")
print(f"IQR: {iqr}")
```

```python
# Question: How can we normalize a dataset using Z-score and what does it look like?

original_data = np.random.normal(80, 10, 200)
z_normalized_data = zscore(original_data)

plt.figure()
sns.kdeplot(original_data, label='Original Data')
sns.kdeplot(z_normalized_data, label='Z-score Normalized')
plt.title("Z-score Normalization")
plt.legend()
plt.show()
```

```python
# Question: What are the skewness and kurtosis of a normally distributed dataset?

data = np.random.normal(0, 1, 1000)
data_skewness = skew(data)
data_kurtosis = kurtosis(data)
print(f"Skewness: {data_skewness}")
print(f"Kurtosis: {data_kurtosis}")
```
